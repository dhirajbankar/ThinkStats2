```python
# Hypothesis using one of the methods covered in Chapter 9.

# So we will test the hypothesis that
# 1. Blood pressure is lower for Mens Patients vs Women Patients. 
#  Hence it is an indicator of high risk for Mortality
# 2. Coleserol is lower for for Mens Patients vs Women Patients.
#  Hence it is an indicator of high risk for Mortality

# To test the hypothesis we will check means of the both
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        #display("1", group1.mean(),"2", group2.mean())
        test_stat = (group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data


# 1. Blood pressue is lower for men Patients vs Women Patients.
# TestStatistics is measuring the value 1 - 2 mean
data =  men.trestbps.values, women.trestbps.values
ht = DiffMeansPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
ht.PlotCdf()
thinkplot.Config(xlabel='test statistic - mean diff for blood pressure', axis=[-10, 30, 0, 1],
                   ylabel='CDF')
```

```python
# 1. Cholesterol is lower for Women Patients vs men Patients.
# TestStatistics is measuring the value 1 - 2 mean
data =  women.chol.values, men.chol.values
ht = DiffMeansPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
ht.PlotCdf()
thinkplot.Config(xlabel='test statistic - mean diff for cholesterol', axis=[-20, 30, 0, 1],
                   ylabel='CDF')
```

```python
def RunTests(heart_df, iters=1000):
    n = len(heart_df)
   
    # Permutation Test
    # cholesterol is lower for men Patients vs women Patients
    data = men.chol.values, women.chol.values
    ht = DiffMeansPermute(data)
    p1 = ht.PValue(iters=iters)
    
    # blood pressure is lower for men Patients vs women Patients
    data = (men.trestbps.values, women.trestbps.values)
    ht = DiffMeansPermute(data)
    p2 = ht.PValue(iters=iters)
    
    print('%d\t%0.2f\t%0.2f' % (n, p1, p2))
```

```python
n = len(heart_df)
for _ in range(7):
    sample = thinkstats2.SampleRows(heart_df, n)
    RunTests(sample)
    n //= 2
```

Conclusion: The tests for test1, test2 are positive throughout. However this pattern may change if we could get large datasets.
```python
# For this project, conduct a regression analysis on either one dependent and one explanatory variable,
# or multiple explanatory variables (Chapter 10 & 11)

# We will try to see the effect on blood pressue by age , chol, exang, oldpeak parameters
# Since the dependent variable is a numaric values, we will use Nonlinear Relationships.
import statsmodels.formula.api as smf
import pandas as pd

formula = 'trestbps ~ age + chol + oldpeak + exang'
model = smf.ols(formula, data=heart_df)
results = model.fit()
results.summary()
# print(model.endog_names)
# print(model.exog_names)
# SummarizeResult(results)
```
